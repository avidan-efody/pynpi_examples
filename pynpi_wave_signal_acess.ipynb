```python
# compile and run the design, nothing very interesting here
import os
from subprocess import Popen, PIPE, CalledProcessError

os.environ['VCS_HOME'] = 'path-to-vcs'
os.environ['VERDI_HOME'] = 'path-to-verdi'

with Popen('simple_example/run', stdout=PIPE, bufsize=1, universal_newlines=True) as p:
    for line in p.stdout:
        print(line, end='')

if p.returncode != 0:
    raise CalledProcessError(p.returncode, p.args)
```

```python
# imports and initialization (i.e. license check)
from pynpi import npisys
from pynpi import waveform

npisys.init(["anything"])
```

```python
# open the relevant fsdb
wave = waveform.open("novas.fsdb")
if not wave:
    print("Error. Failed to open file")
```

```python
# extract some signal values
valid = wave.sig_by_name('top.valid')

print("wave ends at: ", wave.max_time())

valid_values = waveform.sig_hdl_value_between(valid,0,wave.max_time(),waveform.VctFormat_e.DecStrVal)

print(valid_values)
```

```python
# prepare for plotting
import pandas as pd 

df = pd.DataFrame(valid_values, columns =['time', 'value'])
```

```python
# plot the signals, just for fun. we're here for analysis not for competing against Verdi.
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure

output_notebook()

source = ColumnDataSource(df)

p = figure(plot_width=800, plot_height=400, y_range=['1','0'])

p.step(x='time', y='value', line_width=2, source=source)

show(p)
```
